```python
from ecmwf.opendata import Client

client = Client(source="ecmwf")

client.retrieve(
    time=0,
    stream="enfo",
    type="tf",
    step=240,
    target="data.bufr",
)
```

```python
import pdbufr

df = pdbufr.read_bufr("data.bufr",
                      columns=("stormIdentifier", "ensembleMemberNumber", "latitude", "longitude",
             "pressureReducedToMeanSeaLevel", "month", "day", "hour", "minute"))
```

```python
import geopandas as gpd

# SELECTING ONE TROPICAL CYCLONE EXAMPLE:
df = df.query("stormIdentifier == '17E'")
df = df.dropna(subset=['latitude'])

# Convert the Pandas DataFrame to a GeoPandas DataFrame
geometry = gpd.points_from_xy(df['longitude'], df['latitude'])
geopandas_df = gpd.GeoDataFrame(df, geometry=geometry)

# Save the GeoPandas DataFrame to a GeoSpatial file (e.g., GeoJSON)
output_file = "locations.geojson"
geopandas_df.to_file(output_file, driver='GeoJSON')
```

```python
import geopandas as gpd
from shapely.geometry import Point, MultiPolygon
import numpy as np
from rasterio.features import geometry_mask
from rasterio.transform import Affine

transform = Affine(0.1, 0.0, -180.05, 0.0, -0.1, 90.05)

# Assuming geopandas_df is your GeoPandas DataFrame
geopandas_df['geometry'] = geopandas_df.geometry.buffer(1)

lons = np.arange(-180.05, 180, 0.1)
lats = np.arange(90.05, -90, -0.1)

[Lon, Lat] = np.meshgrid(lons, lats)

# Create a numpy array filled with 0s
world_map = np.zeros(np.shape(Lon))

# Iterate through adjacent polygons to create separate convex hulls
for storm in geopandas_df['stormIdentifier'].unique():
    for mem in geopandas_df['ensembleMemberNumber'].unique():
        geopandas_filtered = geopandas_df.query(f"ensembleMemberNumber == {mem} & stormIdentifier == '{storm}'")
        if len(geopandas_filtered) > 0:
            adjacent_polygons = MultiPolygon(list(geopandas_filtered['geometry']))
            convex_hull = adjacent_polygons.convex_hull
            convex_hull_gdf = gpd.GeoDataFrame(geometry=[convex_hull], crs=geopandas_filtered.crs)

            mask = geometry_mask(convex_hull_gdf['geometry'], out_shape=Lon.shape, transform=transform, invert=True)

            world_map = world_map + mask

world_map = world_map / len(geopandas_df['ensembleMemberNumber'].unique())
world_map = np.clip(world_map, 0, 1)
```

```python
import xarray as xr

# Create an xarray DataArray
cyclone_map = xr.DataArray(world_map, coords={'latitude': lats, 'longitude': lons}, dims=['latitude', 'longitude'])
```

```python
import matplotlib.pyplot as plt

cyclone_map.plot()
plt.xlim([-130,-90])
plt.ylim([5,35])
```
